```python
#3
#arithmeticexception occurs due to error in numerical calculations
#There are 3 types of arithmetic_exceptions in python
#a:ZeroDivisionError
#b:OverFlowError
#c:FloatingPointError
def Zero_div_handlling(a,b):
    try:
        c=a/b
    except ZeroDivisionError:
        print("Invalid denominator")
    else:
        print("The ans is :",c)
def Over_flow_handling(a,b):
    try:
        for e in range(b):
            c=a**b
    except OverflowError:
        print("Number too large")
    else:
        print("The ans is :",c)
Zero_div_handlling(4,0) 
Over_flow_handling(3.2,1000)
Zero_div_handlling(4,2) 
Over_flow_handling(3.2,2)
        
    
        
    
```

```python
#4
#Lockup class is used to handle key error and index error
def key_error():
    d={1:2,2:4,3:6}
    try:
        print(d[5])
    except KeyError:
        print("Invalid key")
def index_error():
    a=[1,2,3]
    try:
        print(a[5])
    except IndexError:
        print("Invalid index")
key_error()        
index_error()    
```
